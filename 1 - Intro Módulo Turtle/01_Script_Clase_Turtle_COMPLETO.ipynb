# **2. Configuración de ventana:**

Imagenes MVP Inventech (1).png

```python
# Importación de librería:
import ColabTurtlePlus.Turtle as t

# Borra todo el texto o los dibujos de la pantalla
t.clearscreen()

# Establece el tamaño de la ventana gráfica en píxeles (ancho, alto)
t.setup(200,200)

# Define el fondo del plano de dibujo
t.bgcolor('blue')
```

# **3. Configuración forma de dibujo:**

Para empezar a dibujar con el módulo Turtle(), como lápiz o herramienta de dibujo se puede usar alguna de las formas que la librería tiene incluídas. Estas son algunas de ellas: `'classic', 'turtle', 'circle', 'arrow', 'triangle', 'square', 'ring', 'blank'y 'turtle2'`

```python
t.bgcolor("white")
t.shape("triangle")
```

# **4. Realizar Movimientos:**

Para realizar diferentes movimientos en diferentes direcciones, se pueden usar los siguientes comandos:

*   `forward()`: moverse hacia adelante en px →; *también puedes usar* `fd()`



```python
t.clearscreen()
t.setup(400,50)
t.shape('turtle')

t.fd(100)
```

*   `backward()`: moverse hacia atrás en px ←; *también puedes usar* `bk()`

```python
t.clearscreen()
t.setup(400,50)
t.shape('turtle')

t.bk(100)
```

*   `left()`:  girar a la izquierda en grados; *también puedes usar* `lt()`

```python
t.clearscreen()
t.setup(400,50)
t.shape('turtle')

t.fd(50)
t.left(90)
```

*   `right()`:  moverse hacia la derecha; *también puedes usar* `rt()`

```python
t.clearscreen()
t.setup(400,50)
t.shape('turtle')

t.bk(60)
t.right(90)
```

# **5. Manejo de colores:**

Para cambiar el color de la tortuga y el color de su rastro, se puede usar el comando `color()`, y para cambiar el color del fondo se usa `bgcolor()`. A continuación puedes ver cómo se pueden implementar:

Listado de colores admitidos por los navegadores: https://www.w3schools.com/colors/colors_names.asp

```python
t.clearscreen()
t.setup(300,50)
t.shape('turtle')
t.color('purple')
```

```python
t.color('#29B3DF')
t.bgcolor('#84174E')

t.bk(90)
```

# **6. Rastro de dibujo:**

Para decirle a la tortuga que deje de dibujar se puede usar el comando `up()`, y de esta manera no dejará rastros sobre el plan de trabajo, es decir, si se llega a mover todos sus movimientos pasarán desapercibidos. De esta manera, con el comando `down()` la tortuga vuelve a dejar rastro, y por lo tanto, vuelve a dibujar.

```python
t.clearscreen()
t.setup(300,100)

t.shape('turtle')
t.color('#FF4500')
t.bgcolor("#000080")

t.up()
t.goto(-100,-20)
t.down()
t.fd(200)
```

# **5. Ejercicios prácticos:**



1. Dibujar en un plano de fondo verde un cuadrado de `100 px` de lado de color **blanco**, usando un **triángulo** como forma principal.

Captura de Pantalla 2023-07-25 a la(s) 7.41.14 p. m..png

```python
# Configuración de la ventana
t.clearscreen()
t.setup(200,200)
t.bgcolor("green")
t.color("white")

# Ubicación inicial del dibujo
t.shape("triangle")
t.up()
t.goto(-50,-50)

# Inicio de dibujo
t.left(90)
t.down()
t.fd(100)

t.right(90)
t.fd(100)

t.right(90)
t.fd(100)

t.right(90)
t.fd(100)
```

2. Dibujar un **octágono** cuyos lados sean de `60 px`, usando un fondo azul marino ("**cyan**") y un contorno de color negro. Hazlo utilizando la forma predeterminada de la `tortuga 1`.

Captura de Pantalla 2023-07-25 a la(s) 8.17.11 p. m..png

```python
# Configuración inicial
t.clearscreen()
t.setup(200,200)
t.bgcolor("cyan")
t.shape("turtle")
t.color("black")

# Ubicación del punto inicial para dibujar
t.up()
t.goto(-70,-30)
t.left(90)

t.down()
t.fd(60)
t.right(45)
t.fd(60)

t.right(45)
t.fd(60)

t.right(45)
t.fd(60)

t.right(45)
t.fd(60)

t.right(45)
t.fd(60)

t.right(45)
t.fd(60)

t.right(45)
t.fd(60)
```

3. **¡Dibújalo a tu manera!** Solo debes modificar los colores a tu gusto y cambiar el ángulo de giro en la instrucción `t.left()`:

```python
import ColabTurtlePlus.Turtle as t

t.clearscreen()
t.setup(200,200)

colors = ['red', 'white', 'orange', 'green', 'purple', 'yellow']
t.speed(13)
t.bgcolor('black')

for x in range(181):
	t.color(colors[x%6])
	t.forward(x/2)
	t.left(78)

t.ht()
```
