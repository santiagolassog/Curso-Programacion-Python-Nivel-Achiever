# **2. Condiciones en programación**

**Salida del laberinto:** En el siguiente código la tortuga recorrerá el camino que haya elegido el usuario.

`IMPORTANTE: Analiza el código y comprende cómo funciona`.

```python
import ColabTurtlePlus.Turtle as t

# Configuración de ventana
t.clearscreen()
t.setup(400,400)
t.shape("blank")
t.speed(13)

# Dibujo del laberinto
t.jumpto(-150,150)
t.fd(300)
t.rt(90)
t.fd(300)
t.rt(90)
t.fd(300)
t.rt(90)
t.fd(300)
t.rt(90)

t.jumpto(-150,100)
t.fillcolor("green")
t.begin_fill()
t.fd(50)
t.rt(90)
t.fd(250)
t.rt(90)
t.fd(50)
t.end_fill()

t.jumpto(-50,100)
t.begin_fill()
t.lt(90)
t.fd(200)
t.lt(90)
t.fd(100)
t.lt(90)
t.fd(200)
t.lt(90)
t.fd(100)
t.end_fill()

t.jumpto(150,150)
t.begin_fill()
t.fd(50)
t.lt(90)
t.fd(250)
t.lt(90)
t.fd(50)
t.end_fill()

t.jumpto(115,-170)
t.write(("SAÍDA"), font=("arial", 12, "bold"))

# Ubicación inicial de la tortuga
t.shape("turtle")
t.color("red")
t.jumpto(-150,125)

salida = input("Cuál de los 2 caminos deseas tomar? (1) o (2): ")

# Condiciones del dibujo
if salida == "1":
  t.speed(10)
  t.fd(75)
  t.rt(90)
  t.fd(250)
  t.lt(90)
  t.fd(200)

elif salida == "2":
  t.speed(10)
  t.color("blue")
  t.fd(225)
  t.rt(90)
  t.fd(250)
  t.lt(90)
  t.fd(50)

else:
  t.write(("NO LO SÉ ^_^"), font=("arial", 12, "bold"))
  t.jumpto(-180,125)
```

**4. Yin Yang:** El código que se muestra a continuación te permite dibujar y pintar un `Yin Yang`. Ahora vamos a mejorar el programa permitiéndole al usuario personalizar los colores y preguntarle al usuario cuál de los lados desea mostrar:

1. Mostrar `únicamente` el lado 1
2. Mostrar `únicamente` el lado 2
3. Mostrar ambos lados

`Analiza cuáles serían los ajustes debes hacer al código actual...`

```python
import ColabTurtlePlus.Turtle as t

# Configuración de ventana
t.clearscreen()
t.setup(500,500)
t.shape("turtle")
t.speed(10)

"""-------------------------------------------
            Sección Dibujo Lado 1
-------------------------------------------"""
# Dibujo Lado 1
radius, color1, color2 = 200, 'white', 'black'

t.width(3)
t.pencolor("black")
t.fillcolor(color1)
t.begin_fill()
t.circle(radius/2., 180)
t.circle(radius, 180)
t.left(180)
t.circle(-radius/2., 180)
t.end_fill()
t.fillcolor(color2)
t.left(90)
t.up()
t.forward(radius*0.375)
t.right(90)
t.down()
t.begin_fill()
t.circle(radius*0.125)
t.left(90)
t.end_fill()
t.up()
t.backward(radius*0.375)
t.down()
t.left(90)

"""-------------------------------------------
            Sección Dibujo Lado 2
-------------------------------------------"""
radius, color1, color2 = 200, color2, color1

t.width(3)
t.pencolor("black")
t.fillcolor(color1)
t.begin_fill()
t.circle(radius/2., 180)
t.circle(radius, 180)
t.left(180)
t.circle(-radius/2., 180)
t.end_fill()
t.fillcolor(color2)
t.left(90)
t.up()
t.forward(radius*0.375)
t.right(90)
t.down()
t.begin_fill()
t.circle(radius*0.125)
t.left(90)
t.end_fill()
t.up()
t.backward(radius*0.375)
t.down()
t.left(90)
t.ht()
```
