# **2. Condiciones en programación**

**Salida del laberinto:** En el siguiente código la tortuga recorrerá el camino que haya elegido el usuario.

`IMPORTANTE: Analiza el código y comprende cómo funciona`.

```python
import ColabTurtlePlus.Turtle as t

# Configuración de ventana
t.clearscreen()
t.setup(400,400)
t.shape("blank")
t.speed(13)

# Dibujo del laberinto
t.jumpto(-150,150)
t.fd(300)
t.rt(90)
t.fd(300)
t.rt(90)
t.fd(300)
t.rt(90)
t.fd(300)
t.rt(90)

t.jumpto(-150,100)
t.fillcolor("green")
t.begin_fill()
t.fd(50)
t.rt(90)
t.fd(250)
t.rt(90)
t.fd(50)
t.end_fill()

t.jumpto(-50,100)
t.begin_fill()
t.lt(90)
t.fd(200)
t.lt(90)
t.fd(100)
t.lt(90)
t.fd(200)
t.lt(90)
t.fd(100)
t.end_fill()

t.jumpto(150,150)
t.begin_fill()
t.fd(50)
t.lt(90)
t.fd(250)
t.lt(90)
t.fd(50)
t.end_fill()

t.jumpto(115,-170)
t.write(("SALIDA"), font=("arial", 12, "bold"))

# Ubicación inicial de la tortuga
t.shape("turtle")
t.color("red")
t.jumpto(-150,125)

salida = input("Cuál de los 2 caminos deseas tomar? (1) o (2): ")

# Condiciones del dibujo
if salida == "1":
  t.speed(10)
  t.fd(75)
  t.rt(90)
  t.fd(250)
  t.lt(90)
  t.fd(200)

elif salida == "2":
  t.speed(10)
  t.color("blue")
  t.fd(225)
  t.rt(90)
  t.fd(250)
  t.lt(90)
  t.fd(50)

else:
  t.write(("NO LO SÉ ^_^"), font=("arial", 12, "bold"))
  t.jumpto(-180,125)
```

**3. Cajero electrónico:** Crea un simulador de cajero automático que le permita al usuario retirar dinero de su cuenta digital. Para ello, ten en cuenta los siguientes pasos y criterios:

1. Definir el saldo disponible en la cuenta.
2. Ingresar el monto que se desea retirar.
3. Verificar si hay saldo suficiente.
4. Si el monto deseado es igual o mayor al saldo, permitele al usuario efectuar el retiro.
5. De lo contrario, informale que podrá realizar la transacción pero su saldo será negativo y será compensado en la próxima entrada de dinero a la cuenta; pregúntale si desea efectuar la operación o cancelarla.
6. Mostrar el saldo final una vez sea realizada la transacción.

cajero.png

```python
saldo = 5000

monto = int(input("Cuánto dinero deseas retirar? "))

if monto <= saldo:
  print("Operación realizada con éxito!")
  saldo -= monto

else:
  pregunta = input("Tu saldo es insuficiente. Podrás retirar el dinero, pero en el próximo depósito se realizará la compensación. Deseas continuar? Si/No")

  if pregunta == "Si":
    print("Operación realizada con éxito")
    saldo -= monto
  else:
    print("Operación cancelada")

print("Tu saldo actual es de:",saldo)
```
