# **Introduciendo funciones en programación:**

```python
"""-----------------------------------------
Definición de variables
-----------------------------------------"""
ingredientes = ["lechuga", "tomate", "pepino", "naranja"]

"""-----------------------------------------
Definición/creación de función
-----------------------------------------"""
def preparar_ensalada(ingredientes):

  # Lista de salida
  ensalada = []

  # Ciclo de preparación
  for ingrediente in ingredientes:
      print("Cortar", ingrediente)
      ingrediente = f"{ingrediente} cortad@"
      ensalada.append(ingrediente)

  print("Mezclar ensalada")

  # Retorna/entrega la salida
  return ensalada

"""-----------------------------------------
Uso de la función
-----------------------------------------"""
ensalada_preparada = preparar_ensalada(ingredientes)

print("Ensalada:", ensalada_preparada)
```

# **Variables globales y locales**

**1. Variables Globales:**

Son variables declaradas fuera de cualquier función, generalmente al inicio del script. `Son accesibles desde cualquier parte del código, incluidas las funciones.` Existen durante toda la ejecución del programa, y para modificarlas dentro de una función, se debe usar la palabra `global`.

```python
x = 10  # Variable global

def imprimir_x():
    print(x)  # Accede a la variable global

def modificar_x():
    global x
    x = 20    # Modifica la variable global

imprimir_x()  # Salida: 10
modificar_x()
imprimir_x()  # Salida: 20
```

**2. Variables Locales:**

Son variables declaradas dentro de una función. `Solo son accesibles dentro de la función en la que se declaran.` Existen solo durante la ejecución de la función, y no afectan ni son afectadas por las variables globales con el mismo nombre.

```python
def mostrar_valor():
  # Variable local
    y = 5
    print("Variable local:", y)

# Variable global
y = 10
print("Variable global:", y)
mostrar_valor()
```

# **Mezcla de colores**

Captura de Pantalla 2024-06-25 a la(s) 4.56.02 p. m..png

Puedes usar el siguiente [mezclador de colores online](https://gradients.app/es/mix) para que puedas experimentar con colores personalizados.

```python
# Configuración de ventana
t.clearscreen()
t.setup(250,50)
t.speed(13)

# Colores a mezclar
color1 = (255,255,0) # Amarillo
color2 = (0,0,255)   # Azul
color3 = (255,0,0)   # Rojo

# Función que dibuja los círculos
def caminar(color):
  t.color(color)
  t.begin_fill()
  t.circle(20)
  t.end_fill()
  t.fd(50)
  t.up()

# Función que mezcla los colores
def mezclar_colores(c1, c2, c3):
  r = (c1[0] + c2[0] + c3[0]) // 3
  g = (c1[1] + c2[1] + c3[1]) // 3
  b = (c1[2] + c2[2] + c3[2]) // 3

  color_mezclado = (r, g, b)
  print(f"Color mezclado: RGB: {color_mezclado}")

  return color_mezclado

# Inicio de dibujo
t.jumpto(-75,-20)
caminar(color1)
caminar(color2)
caminar(color3)

# Mezcla de colores
color_mezclado = mezclar_colores(color1, color2, color3)
caminar(color_mezclado)

t.ht()
```
