# **3. Introduciendo el Ciclo While**

GIF While Loop.gif

```python
import time

meta = 0

while meta < 4:

    # Proceso que se repite
    print("Avanzar")
    print("Girar a la izquierda")
    print("Avanzar")
    print("Girar a la derecha")
    print("--------------")

    # Aumenta en 1 paso
    meta += 1

    print("Valor:", meta)
    time.sleep(1)

print("¡Fin del ciclo!")
```

# **Carrera Hot Wheels:**

**Carrera Hot Wheels:** Este código simula una emocionante carrera entre dos autos. Utiliza un `ciclo while` para lograr que los corredores avancen a una distancia aleatoria en cada `iteración`. La carrera continúa hasta que uno de los corredores alcanza la meta. ¿Quién será el ganador? ¡Descúbrelo ejecutando el programa!"

hot-wheels-hot-wheels-jump.gif

```python
import random
import time

# Distancia total de la carrera
distancia_meta = 100

# Posiciones iniciales de los corredores
pos_car1 = 0
pos_car2 = 0

# Simulación de la carrera
while pos_car1 < distancia_meta and pos_car2 < distancia_meta:

    # Genera la distancia a recorrer
    dist1 = random.randint(1,20)
    dist2 = random.randint(1,20)

    # Avance de los corredores
    pos_car1 = pos_car1 + dist1  # El carro 1 avanza "dist1" unidades
    pos_car2 = pos_car2 + dist2  # El carro 2 avanza "dist2" unidades

    # Mostrar las posiciones de los corredores
    print("Posición del carro 1:", pos_car1)
    print("Posición del carro 2:", pos_car2)
    print("-----------------------------")

    # Genera un retardo de 1 segundo
    time.sleep(2)

# Determina el ganador de la carrera
if pos_car1 >= distancia_meta:
    print("¡El corredor 1 ha ganado!")
else:
    print("¡El corredor 2 ha ganado!")
```

# **Ejercicio Práctico: Configuración**

```python
def config():

  # Screen setup
  clearscreen()
  setup(800,500)
  bgcolor("#EFEFEF")
  speed(0)

  # Heading
  penup()
  goto(0,215)
  color("black")
  write("CARRERA DE TORTUGAS", align="center", font=("arial", 20, "bold"))

  # Dirt
  goto(-350,200)
  pendown()
  color("grey")
  begin_fill()

  for i in range(2):
    forward(700)
    right(90)
    forward(400)
    right(90)
  end_fill()

  # Finish Line
  gap_size = 20
  shape("square")
  penup()
  color("white")

  for i in range(10):
    goto(250,(170-(i*gap_size*2)))
    stamp(layer=1)

  for i in range(10):
    goto(250+gap_size,(210-gap_size-(i*gap_size*2)))
    stamp(layer=1)

  color("black")

  for i in range(10):
    goto(250,(190-(i*gap_size*2)))
    stamp(layer=1)

  for i in range(10):
    goto(251+gap_size,(190-gap_size-(i*gap_size*2)))
    stamp(layer=1)

  done()
```

**Paso 2: Movimiento de jugadores**

La distancia que cada tortuga debe recorrer, es igual a `650 píxeles`.

1. Para determinar la distancia recorrida por la tortuga, necesitamos crear variables que tendrán un valor inicial de 0. Vamos a nombrarlas: `t1_dist`, `t2 _dist` y `t3_dist`.

2. Toda la carrera se programará dentro de un `bucle while`. El bucle se ejecutará `hasta que` una de las tortugas llegue a la línea de meta.

3. En cada iteración, usando la función `random.randint(1, 10)`, determinaremos la distancia que cada tortuga va a recorrer, y las vamos a mover con ese valor.

4. Recuerda `acumular` el valor de pasos recorridos por cada tortuga, usando las variables previamente creadas.

```python
import random

# Interfaz gráfica del juego
config()

# Crear los jugadores
t1 = Turtle()
t2 = Turtle()
t3 = Turtle()

# Definir formas de tortuga
t1.shape("turtle")
t2.shape("turtle")
t3.shape("turtle")

# Colores de cada tortuga
t1.color("red")
t2.color("blue")
t3.color("yellow")

# Posiciones iniciales de cada tortuga
t1.jumpto(-380,100)
t2.jumpto(-380,0)
t3.jumpto(-380,-100)

# Variables de posición de jugadores
t1_dist = 0
t2_dist = 0
t3_dist = 0

# Bucle principal
while True:

    if t1_dist > 650 or t2_dist > 650 or t3_dist > 650:
      break

    dist1 = random.randint(1, 10)
    dist2 = random.randint(1, 10)
    dist3 = random.randint(1, 10)

    # Movimiento de jugadores
    t1.fd(dist1)
    t2.fd(dist2)
    t3.fd(dist3)

    # Acumular valores de recorrido
    t1_dist += dist1
    t2_dist += dist2
    t3_dist += dist3
```

**Paso 3: Definir al ganador**

Una vez finalizada la carrera guardaremos el color de la tortuga que ganó la carrera en la variable `winner`. Y finalmente muestra en consola un mensaje que muestre cuál fue la tortuga ganadora.

```python
import random

# Interfaz gráfica del juego
config()

# Crear los jugadores
t1 = Turtle()
t2 = Turtle()
t3 = Turtle()

# Definir formas de tortuga
t1.shape("turtle")
t2.shape("turtle")
t3.shape("turtle")

# Colores de cada tortuga
t1.color("red")
t2.color("blue")
t3.color("yellow")

# Posiciones iniciales de cada tortuga
t1.jumpto(-380,100)
t2.jumpto(-380,0)
t3.jumpto(-380,-100)

# Variables de posición de jugadores
t1_dist = 0
t2_dist = 0
t3_dist = 0

# Bucle principal
while True:

    if t1_dist > 650 or t2_dist > 650 or t3_dist > 650:
      break

    dist1 = random.randint(1, 10)
    dist2 = random.randint(1, 10)
    dist3 = random.randint(1, 10)

    # Movimiento de jugadores
    t1.fd(dist1)
    t2.fd(dist2)
    t3.fd(dist3)

    # Acumular valores de recorrido
    t1_dist += dist1
    t2_dist += dist2
    t3_dist += dist3

# Definir al ganador
winner = max(t1_dist, t2_dist, t3_dist)

if t1_dist == winner:
  print("La tortuga ganadora fue la ROJA")
elif t2_dist == winner:
  print("La tortuga ganadora fue la AZUL")
elif t3_dist == winner:
  print("La tortuga ganadora fue la AMARILLA")
```
