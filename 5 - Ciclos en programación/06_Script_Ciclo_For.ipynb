# **Introduciendo el Ciclo For:**

En este ejemplo, simulamos el conteo de pasos de un pingüino durante 5 segundos. Utilizamos un `ciclo for` para iterar sobre cada segundo y obtener la cantidad de pasos de nuestro pequeño amigo. Al final, imprimimos el total de pasos caminados en los 5 segundos.

pasos.gif


```python
import time
import random

# Variable que cuenta los pasos mientras totales
pasos_totales = 0

print("Inicio del conteo...🐧🐧🐧")
print("-------------------------------------------------------------")

# Simulación de pasos durante 5 segundos
for i in range(5):
    pasos = random.randint(10,20)
    time.sleep(1)
    print(f"Pasos caminados en el segundo {i+1}: {pasos}")
    pasos_totales += pasos

print("-------------------------------------------------------------")
print(f"Has caminado un total de {pasos_totales} pasos en 5 segundos.")
```

# **Estrella de 9 puntas:**

```python
import time

# Configuración de ventana
t.clearscreen()
t.setup(300,300)

# Colores
fondo = '#040c5c'
color_dibujo = 'white'
color_contorno = 'yellow'
t.bgcolor(fondo)
t.color(color_dibujo)
t.pencolor(color_contorno )

# Configuración de dibujo
t.speed(13)
t.width(0.5)
t.speed(13)
t.jumpto(100,15)

# Variables de dibujo
angle = 160
iteraciones = 10

# Dibujo de la estrella
t.begin_fill()

for i in range(iteraciones):
    t.forward(-200)
    t.left(angle)
    time.sleep(1)

t.ht()
t.end_fill()
```
