# **CRUD en listas:**

Presentación Curso Programación Achiever - FREE.png

**"CRUD"** es un acrónimo que representa las cuatro operaciones básicas que se pueden realizar en una base de datos o en una estructura de datos como una lista. Estas operaciones son:

1. **Create** (Crear): Crear una lista de nuevos elementos.
2. **Read** (Leer): Leer o acceder a los datos existentes.
3. **Update** (Actualizar): Modificar los datos existentes.
4. **Delete** (Eliminar): Borrar los datos existentes

````python
# 1. CREATE (Crear Lista)
videojuegos = ["Minecraft", "Fortnite", "Among Us", "Call of Duty"]

# 2. READ (Leer Lista): Imprimir la lista completa de videojuegos
print("Lista de videojuegos:", videojuegos)

# 2.1 Leer el nombre de un videojuego específico
print("Primer videojuego:", videojuegos[0])
print("Último videojuego:", videojuegos[-1])

# 3. UPDATE (Actualizar Lista): Cambiar el nombre de un videojuego
videojuegos[2] = "Grand Theft Auto"
print("Lista de videojuegos actualizada:", videojuegos)

# 4. DELETE (Eliminar Elementos en Lista): Eliminar un videojuego por su nombre
videojuegos.remove("Minecraft")
print("Lista después de eliminar 'Minecraft':", videojuegos)

# 4.1 Eliminar un videojuego por su índice
del videojuegos[-1]  # Elimina 'Call of Duty', que está en la última posición
print("Lista después de eliminar 'Call of Duty':", videojuegos)
````

# **Recorrido de listas:**

image.png

En este ejercicio podrás ver `2 formas` de acceder a los elementos de una lista, utilizando el `ciclo for`. Con ayuda de la `variable de control`, es súper sencillo acceder a la información y tomar decisiones. Para ello, utilizar el siguiente `compilador y depurador` online para ver cómo funciona el siguiente código: **https://pythontutor.com/python-compiler.html#mode=edit**

````python
"""--------------------------------------------
# OPCIÓN 1: FORMA CLÁSICA --> for i in range()
--------------------------------------------"""
inventario = ["manzanas", "naranjas", "peras", "bananos"]

# Mostrar el inventario
print("Inventario actual:", inventario)

# Nuevos productos que llegaron
nuevos_productos = ["uvas", "kiwis"]

# Número de productos en la lista
cantidad = len(nuevos_productos)

# Agregar nuevos productos al inventario
for producto in range(cantidad):
    print(producto)
    inventario.append(nuevos_productos[producto])

print("Inventario final:",inventario)
````

````python
"""--------------------------------------------
# OPCIÓN 2: FORMA DIRECTA --> for i in lista:
--------------------------------------------"""

inventario = ["manzanas", "naranjas", "peras", "bananos"]

# Mostrar el inventario
print("Inventario actual:", inventario)

# Nuevos productos que llegaron
nuevos_productos = ["uvas", "kiwis"]

# Agregar nuevos productos al inventario
for producto in nuevos_productos:
    print(producto)
    inventario.append(producto)

print("Inventario final:",inventario)
````

# **Lista de amigos:**

istockphoto-1214380596-612x612.jpg

En este ejercicio se emplea el módulo Turtle para mostrar de forma visual el nombre de cada uno de tus amigos. Para ello, el programa te solicita que ingreses el nombre de cada amigo, y posteriormente lo muestra en la interfaz con un color diferente. Analiza el código de principio a fin, y descubre cómo son empleadas las listas en el ejercicio.

````python
import random

# Configuración de ventana
t.clearscreen()
t.setup(450,50)
t.shape('turtle')

# Colores de dibujo
fondo = 'white'
t.bgcolor(fondo)

# Crear la lista de amigos
amigos = []
cant_amigos = 4

# Posición inicial
t.jumpto(-170,-5)

# Ciclo que muestra los nombres de la lista de amigos
for i in range(cant_amigos):

  nombre = input(f'Ingresa el nombre de tu amigo {i+1}: ')
  # Agrega el nombre del amigo a la lista de amigos
  amigos.append(nombre)

  # Genera colores al azar
  r = random.randint(0,255)
  g = random.randint(0,255)
  b = random.randint(0,255)
  t.color(r,g,b)

  # Escribe el nombre de cada amigo de acuerdo a la posición
  t.write(f'{i+1}) {amigos[i]}', font=("arial", 15, "bold"))

  # Realiza los movimientos
  t.fd(60)
  t.up()
  t.fd(30)
  t.down()

t.ht()
````

# **Ejercicio Práctico: Lista de Tareas Diarias**

the-pros-and-cons-of-paper-to-do-lists.jpeg

Vamos a desarrollar un programa que te ayudará a mantener un seguimiento de tus tareas diarias. Este programa te permitirá ver qué tareas tienes pendientes, agregar nuevas tareas a tu lista, marcar tareas como completadas, y finalizar el programa cuando ya no necesites hacer más cambios.

**Paso a paso de implementación:**

1. Crear una lista de tareas: Empezaremos con una lista predefinida de tareas.

```
tareas = [
    "Hacer ejercicio",
    "Comprar alimentos",
    "Estudiar programación",
    "Llamar a mis padres",
    "Leer un libro"
]
```

2. Definir las acciones del programa: Vamos a definir las diferentes acciones que nuestro programa puede realizar. Estas acciones son las opciones que el usuario podrá elegir.

```
acciones = ['1) Mostrar tareas pendientes',
            '2) Agregar una nueva tarea',
            '3) Marcar una tarea como completada',
            '4) Finalizar programa']
```
3. Crear un bucle para el menú de acciones: Utilizaremos un bucle while para mostrar continuamente las acciones hasta que el usuario decida finalizar el programa. Dentro del bucle, mostraremos el menú y pediremos al usuario que elija una opción.

4. Implementar las Funcionalidades: Dependiendo de la opción que elija el usuario, el programa realizará las diferentes acciones.

Este ejercicio te ayudará a gestionar tus propias tareas diarias de manera más eficiente y te permitirá practicar cómo manejar listas, ciclos y condicionales en un contexto que es relevante para tu vida diaria.


````python
# Lista de tareas
tareas = [
    "Hacer ejercicio",
    "Comprar alimentos",
    "Estudiar programación",
    "Llamar a mis padres",
    "Leer un libro"
]

# Lista de acciones
acciones = ['1) Mostrar tareas pendientes',
            '2) Agregar una nueva tarea',
            '3) Marcar una tarea como completada',
            '4) Finalizar programa']

"""-----------------------------------------------------------------
Inicio del programa
-----------------------------------------------------------------"""
while True:

    # Mostrar lista de opciones/acciones
    for opcion in acciones:
      print(opcion)

    # Preguntar al usuario por la opción que desea
    opcion_elegida = int(input("Cuál es la opción que deseas elegir? "))

    """-----------------------------------------------------------------
    Codificación de funcionalidades del programa
    -----------------------------------------------------------------"""

    # Mostrar tareas pendientes
    if opcion_elegida == 1:

      for tarea in tareas:
        print(tarea)

    # Agregar una nueva tarea
    elif opcion_elegida == 2:

      nueva_tarea = input("Cuál es la tarea que deseas agregar? ")
      tareas.append(nueva_tarea)

    # Marcar una tarea como completada
    elif opcion_elegida == 3:

      tarea_completada = input("Cuál es la tarea que deseas marcar como completada? ")
      if tarea_completada in tareas:
        tareas.remove(tarea_completada)

    # Finalizar el programa
    elif opcion_elegida == 4:
      print("Gracias por utilizar el programa.")
      break
````
